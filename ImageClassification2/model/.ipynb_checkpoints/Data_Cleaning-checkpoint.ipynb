Haarcascades helps us to detect each features of a face

```python
face_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('./opencv/haarcascades/haarcascade_eye.xml')

faces = face_cascade.detectMultiScale(gray_img, 1.3, 5)
faces
```

```python
(x,y,w,h) = faces[0] # store the x, y, width and height 
x,y,w,h
```

```python
# draw a rectangle around the place that it detected
face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(face_img)
```

```python
cv2.destroyAllWindows()
for (x,y,w,h) in faces: # loop through the faces present
    face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    # roi = region of interest, here the faces
    roi_gray = gray_img[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w] 
    # take out each face and then use eye cascade on each
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes: # loop through the eyes detected
        # draw rectangle around each eye detected
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        

plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
```

So using the above techniques, let's write a function that would mark out face and eyes if visible in the photo

```python
def get_cropped_image_if_2_eyes(image_path):
    img = cv2.imread(image_path) # read image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # convert to gray image
    faces = face_cascade.detectMultiScale(gray, 1.3, 5) # detect face
    for (x,y,w,h) in faces:
        # region of interest taken
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray) # detect eyes
        # if a photo has >= 2 eyes then only return the roi face
        if len(eyes) >= 2:
            return roi_color
```

```python
original_image = cv2.imread('./test_images/israel.jpg')
plt.imshow(original_image)
```

```python
cropped_image = get_cropped_image_if_2_eyes('./test_images/israel.jpg')
plt.imshow(cropped_image)
```

Now we'll create a folder to store the cropped photos like the above one in.

```python
path_to_data = "./dataset"
path_to_cr_data = "./dataset/cropped/"
```

```python
import os
img_dirs = []
# scan all the directories present in path_to_data
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
```

```python
img_dirs
```

```python
import shutil
# check if a folder exists, if does then remove it
if os.path.exists(path_to_cr_data):
    shutil.rmtree(path_to_cr_data)
os.mkdir(path_to_cr_data) # create a folder
```

create cropped folders for each celeb. scan images from original folder, crop them using the cropped function and store them in respective cropped folder

```python
cropped_image_dirs = []
celebs_file_names_dict = {}

# scan each celeb folder
for img_dir in img_dirs:
    count = 1
    celeb_name = img_dir.split('\\')[-1] # get each folders name
    print(celeb_name)

    celebs_file_names_dict[celeb_name] = []
    
    # scan each image in the folder
    for entry in os.scandir(img_dir):
        roi_color = get_cropped_image_if_2_eyes(entry.path) # crop the image
        if roi_color is not None: # if face visible properly
            # create the folder path
            cropped_folder = path_to_cr_data + celeb_name
            # if such folder doesn't exists, create one with the above path
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder) # keep track
                print("Generating cropped images in folder: ", cropped_folder)

            # save the cropped images as png, name like conor1, conor2, etc
            cropped_file_name = celeb_name + str(count) + ".png"
            cropped_file_path = cropped_folder + "/" + cropped_file_name
            # save the file into the folder created
            cv2.imwrite(cropped_file_path, roi_color)
            celebs_file_names_dict[celeb_name].append(cropped_file_path)
            count += 1
    
    
```

#### Now we will convert the image into such a one that is very useful to a computer using py wavelet transform (uses fourier transform)

```python
import numpy as np
import pywt
import cv2
```

```python
def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype Conversions
    #Convert to grayscale
    imArray = cv2.cvtColor(imArray, cv2.COLOR_RGB2GRAY)
    #convert to float
    imArray = np.float32(imArray)
    imArray /= 255
    #compute coeffs
    coeffs = pywt.wavedec2(imArray, mode, level=level)

    #Process coeffs
    coeffs_H = list(coeffs)
    coeffs_H[0] *= 0

    #reconstruction
    imArray_H = pywt.waverec2(coeffs_H, mode)
    imArray_H *= 255
    imArray_H = np.uint8(imArray_H)

    return imArray_H
```

```python
im_har = w2d(cropped_image, 'db1', 5)
plt.imshow(im_har, cmap='gray')
```

#### This is a wavelet transform image. Gives the computer every minute details of the face -- how big or small the eyes are, where are the eyes placed, how are the lips, size and structure and placement of the nose, etc

```python
# stores celeb name and correspondigly their images path in an array
celebs_file_names_dict 
```

```python
# Let's give a random number to each celeb, so that our model can classify
class_dict = {}
count = 0
for celeb_name in celebs_file_names_dict.keys():
    class_dict[celeb_name] = count
    count += 1
class_dict    
    
```

Now let's stack both gray and wavelet transform image vertically and that stack will be the input to our classifier. 

```python
X, y = [], []
for celeb_name, training_files in celebs_file_names_dict.items():
    for training_image in training_files:
        img = cv2.imread(training_image)
        if img is None:
            continue
        scalled_raw_img = cv2.resize(img, (32, 32)) # scale to 32 by 32 pixel
        img_har = w2d(img, 'db1', 5) # get the wavelet transform image
        scalled_img_har = cv2.resize(img_har, (32, 32)) # scale that too
        # Now stack raw image on top of 
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3, 1), scalled_img_har.reshape(32*32, 1)))
        X.append(combined_img)
        y.append(class_dict[celeb_name])
```
